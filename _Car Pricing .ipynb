```python
sns.lineplot(x='Car_Age',y='Depreciation',data=df)
plt.title('Car age vs depriciation')
plt.show()
```

```python
summary = df.groupby('Fuel_Type')[['Selling_Price','Depreciation','kms_driven_per_year']].mean()
summary
```

```python
#Avg depreciation rate
df['Depreciation%'] = np.round((df['Depreciation']/df['Price'])*100,2)
df
```

```python
avg_dep = np.mean(df['Depreciation%'])
avg_dep
```

```python
#Insight
Dep rate is of vehicle is 36
```

```python
df['dep per year']  = df['Depreciation%'] / df['Car_Age']
df
```

```python
sns.lineplot(y='dep per year', x='Fuel_Type',data=df)
plt.title('fuel type vs dep per year')
plt.show()
```

```python
Insight
* Diesel vehicle is having less depreciation per year as compared to petrol and cng
```
